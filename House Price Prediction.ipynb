# House_Price_Prediction

```python
import pandas as pd
import numpy as np
```

```python
dataoriginal=pd.read_csv("Bengaluru_House_Data.csv")
dataoriginal
```

```python
dataoriginal.drop(columns=['area_type','availability','society'],inplace=True)
```

```python
data=dataoriginal
data.info()
```

# Data Preprocessing

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.location.value_counts()
```

```python
# making the data locations with no spaces
data['location']=data['location'].apply(lambda x:x.strip())
data.location.value_counts()
```

```python
location_stats = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_stats
```

```python
location_with_10_entries=location_stats[location_stats <=10]
location_with_10_entries
```

```python
# eliminating the locations <=10 entries
data['location']=data['location'].apply(lambda x:'other' if x in location_with_10_entries else x)
data.location.value_counts()
```

```python
data
```

```python
data['size'].unique()
```

```python
data['bedrooms']=data['size'].apply(lambda x:int(x.split(' ')[0]))
```

```python
data
```

```python
data.total_sqft.unique()
```

```python
#For making sqft as float and avg if prsent in range
def clean(sqft):
    tokens = sqft.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+ float(tokens[1]))/2
    else:
        try:
            return float(sqft)
        except:
            return None
```

```python
data['total_sqft']=data['total_sqft'].apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
#No Null Values are present now in our data
```

```python
#If any outliers are present then should remove them
data['sqft_per_bed']= data['total_sqft']/data['bedrooms']
```

```python
data.sqft_per_bed.describe()
```

```python
data= data[data['sqft_per_bed']>=300]
```

```python
data
```

```python
# Use .loc[] when slicing the DataFrame to avoid the warning
data1 = data.loc[:, ['location','size','total_sqft','bath','balcony','bedrooms','price','sqft_per_bed']]

# Now safely add the new column using .loc[]
data1.loc[:, 'price_per_sqft'] = data1['price'] * 100000 / data1['total_sqft']
```

```python
data1.price_per_sqft.describe()
```

```python
data2=data1[data1['price_per_sqft']>=2000]
```

```python
data2
```

```python
# List of columns you want to drop
data3=data2.drop(columns=['size', 'sqft_per_bed', 'price_per_sqft'],axis=1)
```

```python
data3
```

```python
#Transformation of data using oneHot Encoder
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse=False), ['location']), remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
scaler=StandardScaler()
```

```python
model=make_pipeline(col_trans,scaler,lr)
```

```python
data_input=data3.drop(columns=['price'])
data_output=data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
input=pd.DataFrame([['Electronic City Phase II',1700.0,2.0,2.0,3.0]],columns=['location','total_sqft','bath','balcony','bedrooms'])
```

```python
model.predict(input)
```

```python
import pickle as pk
```

```python
pk.dump(model,open('House_prediction_model.pkl','wb'))
```
